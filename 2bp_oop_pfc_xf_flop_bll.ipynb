## Paso 1 — Distribución de `Flop Made Hand`

**Pregunta:** ¿Qué tipos de manos estoy foldeando con más frecuencia?

**Valor estratégico:** Este spot solo contiene XF. Si foldeamos mucho con categorías que suelen tener equity (pares, overcards con backdoors, etc.), es evidencia de posible over-folding vs small c-bet.

```python
flop_hand_counts = df["Flop Made Hand"].value_counts(dropna=False)
flop_hand_pct = df["Flop Made Hand"].value_counts(normalize=True).mul(100).round(1)

flop_hand_summary = pd.DataFrame({
    "count": flop_hand_counts,
    "pct": flop_hand_pct,
})

print(f"Total manos: {len(df)}")
print()
flop_hand_summary
```
